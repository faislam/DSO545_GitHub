```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#1. Density plots

data = pd.read_csv('mpg.csv')
data.head()
```

```python
plt.figure()

sns.kdeplot(data.loc[data.cyl == 4, 'hwy'],
            shade = True,
            color = 'green',
            label = 'Cyl = 4',
            alpha = 0.4
            )
sns.kdeplot(data.loc[data.cyl == 5, 'hwy'],
            shade = True,
            color = 'deeppink',
            label = 'Cyl = 5',
            alpha = 0.4
            )
sns.kdeplot(data.loc[data.cyl == 6, 'hwy'],
            shade = True,
            color = 'dodgerblue',
            label = 'Cyl = 6',
            alpha = 0.4
            )
sns.kdeplot(data.loc[data.cyl == 8, 'hwy'],
            shade = True,
            color = 'orange',
            label = 'Cyl = 8',
            alpha = 0.4
            )
plt.show()
```

```python
#2. Create a 2x3 grid

plt.figure(figsize = (10,5))

plt.subplots_adjust(hspace = 0.5, wspace = 0.6) #hspace = height space wspace = widthspace

for i in range(1,7):
    plt.subplot(2,3,i) #creates 6 different subplots
    plt.text(0.5,0.5, str((2,3,i)),ha = 'center') #ha: horizontal alignment, centers text

plt.show()
```

```python
#3. Create a 2x3 graph with sin and cos plots (different approach from #2)

plt.figure()

fig,ax = plt.subplots(2,3, sharex = 'col', sharey = 'row')

x = np.linspace(0,50)

ax[0,0].plot(np.sin(x))
ax[0,2].plot(np.cos(x), color = 'red')

plt.show()
```

```python
#4.

fig = plt.figure()

grid = plt.GridSpec(2,3,wspace = 0.4, hspace = 0.5)

fig.add_subplot(grid[0,0])
fig.add_subplot(grid[0,1:3])
fig.add_subplot(grid[1,0:2])
fig.add_subplot(grid[1,2])

plt.show()
```

```python
#5. Marginal Boxplots

fig = plt.figure(figsize = (10,5))

grid = plt.GridSpec(4,4, wspace = 0.4, hspace = 0.5)

ax_main = fig.add_subplot(grid[:3,:3]) #OR grid[0:-1,:3] goes all the way to one row before end
ax_right = fig.add_subplot(grid[:3,3])
ax_bottom = fig.add_subplot(grid[3,:3])

## plot the main graph (scatterplot)
ax_main.scatter(data.displ, data.hwy)

## plot the right boxplot
sns.boxplot(data.hwy, ax = ax_right, orient = 'v')
ax_right.set(xticks = [], yticks = [], ylabel = '')

## plot the bottom boxplot
sns.boxplot(data.displ, ax = ax_bottom, orient = 'h')
ax_bottom.set(xticks = [], yticks = [], xlabel = '')

## set title, xlabel, ylabel
ax_main.set(title = 'Scatterplot with Boxplots \n displacement v. highway mileage',
            xlabel = 'Displacement',
            ylabel = 'Highway Mileage'
           )

plt.show()
```

```python
#6. 

data = pd.read_csv('economics.csv')
data.head()
```

```python
fig = plt.figure(figsize = (12,5))

plt.plot(data.date, data.psavert, color = 'tab:red')
ax = plt.gca()

## set main y axis (left)
ax.tick_params(axis = 'y', labelcolor = 'tab:red')
ax.set_ylabel('Personal Savings Rate', color = 'tab:red', fontsize = 14)

## set the secondary y axis (right)
ax_right = ax.twinx() ## create a twin y axis that shares the same x axis
ax_right.plot(data.date, data.unemploy, color = 'tab:blue')
ax_right.tick_params(axis = 'y', labelcolor = 'tab:blue')
ax_right.set_ylabel("# Unemployed(1000's)", color = 'tab:blue', fontsize = 14)

## set the x axis
ax.set_xticks(np.arange(0,len(data.date),60))

plt.title('Personal Savings Rate vs. Unemployment', fontsize = 20)

plt.show()
```
